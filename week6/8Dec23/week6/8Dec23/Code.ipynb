```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
def imshow(image, title=None, size=10):
    assert image is not None, 'Not Provided'
    plt.imshow(image, 'gray')
    plt.title(title)
    plt.axis('off')
    plt.show()
```

```python
image = cv2.imread('./../newImages/Fig0304(a)(breast_digital_Xray).tif', 0)
```

```python
imshow(image, 'InputImage')
```

```python
negativeImage = 255-image
imshow(image, 'InputImage')
imshow(negativeImage, 'NegativeImage')
```

```python
negativeImage = np.max(image)-image
imshow(image, 'InputImage')
imshow(negativeImage, 'NegativeImage')
```

```python
# outPut = Asx(f(x,y), - msxy)+msxy
# asx = k*mg/sxy
image = cv2.imread('./../newImages/Fig0326(a)(embedded_square_noisy_512).tif', 0)
result = np.zeros((512,512))
gmean = np.mean(image)
k = 1
for i in range(512):
    for j in range(512):
#         print(i,j)
#         print(image[i,j])
#         print(image[i:i+5,j:j+5])
        region = image[i:i+5,j:j+5]
        smean = np.mean(region)
        stdRegion = np.std(region)
        lgf = (k*gmean)/stdRegion
        result[i,j] = (lgf * (image[i,j] - smean))+smean
imshow(image, 'InputImage')
imshow(result, 'ResultImage')
```
